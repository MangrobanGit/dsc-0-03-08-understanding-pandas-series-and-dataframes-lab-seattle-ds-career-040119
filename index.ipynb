## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
station_cts = df['station'].value_counts().to_frame().reset_index()
station_cts.columns = ['station','Num_Lines']
station_cts.head()
```

## Write a function to clean a column name.

```python
def clean(col_name):
    #cleaned = #Your code here; whatever you want to do to col_name. Hint: think back to str methods.
    # Strip out whitespace
    cleaned = col_name.lower().strip()
    return cleaned
```

```python
#90	08/25/2018	00-00-00	42 ST-PORT AUTH	IND	ACENQRS1237W	1622021	3170464	1548443

df_new[ ( df_new['division'] == 'IND') & (df_new['date'] == '08/25/2018') &
       ( df_new['scp'] == '00-00-00') & (df_new['linename'] == 'ACENQRS1237W')].head(20)
```

```python
df_new.info()
```

```python
#df_new[ df_new['station'] == '111 ST'].groupby(['date','station'], as_index = False ).agg(
#    { "exits": "min" } ).rename(columns={'exits':'exits_max'})
exits_min_df = df_new.groupby(['date','scp','station','division','linename','unit'], as_index = False ).agg(
    { "exits": "min" } ).rename(columns={'exits':'exits_min'})
exits_min_df[ (exits_min_df['station'] == '111 ST') & (exits_min_df['scp'] == '00-00-00') 
             & ( exits_min_df['division'] == 'BMT')]
```

```python
exits_max_df = df_new.groupby(['date','scp','station','division','linename','unit'], as_index = False ).agg(
    { "exits": "max" } ).rename(columns={'exits':'exits_max'})
exits_max_df[ ( exits_max_df['station'] == '111 ST') & (exits_max_df['scp'] == '00-00-00') 
             & (exits_max_df['division'] == 'BMT')]
```

Ok, I think it is a combination of `scp` | `station` | `division`|`unit`|`linename` ( and `date` )

```python
exits_df = pd.merge( exits_min_df, exits_max_df
                   ,how = 'inner'
                   ,left_on = ['date','scp','station','division','linename','unit']
                   ,right_on = ['date','scp','station','division','linename','unit'] 
                   )
exits_df.head(5)
# 2	08/25/2018	00-00-00	103 ST	IRT	518211	505305968
```

Minor cleanup

Create column `exits_ct` that will assume to be difference in of max and min

```python
exits_df['exits_ct'] = exits_df['exits_max'] - exits_df['exits_min']
exits_df.head(5)
```

Create day of the week column `dotw`

```python
# first, date is actually not datetime
exits_df['date'] = pd.to_datetime(exits_df['date'])

# Extract DOTW
exits_df['dotw'] = exits_df['date'].dt.strftime('%A')
exits_df.head()
```

```python
# Group by DOTW
dotw_smry = exits_df.groupby(['dotw']
                            ,as_index = False).agg({"exits_ct":"sum"})
dotw_smry
```

##### Finally, show chart!

```python
import numpy as np 

fig, ax = plt.subplots(figsize=(5,5))

# Get data
# Sort it first
dotw_smry.sort_values( by = ['exits_ct'], inplace = True)

# going with horizontal bar, so:
width = dotw_smry['exits_ct']
y_pos = np.arange(len(dotw_smry['dotw']))
y_labels = dotw_smry['dotw']

# Select Plot
ax.barh( y = y_pos
       ,width = width
       ,align = 'center' )

# Set axes labels and aesthetics
ax.set_ylabel('Day of the Week')
ax.set_xlabel('Turnstile Traffic')
ax.set_yticks(y_pos)
ax.set_yticklabels(y_labels)
ax.invert_yaxis

plt.show()
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

On visual inspection of above, there's way less traffic on a weekend. How to compare?

Let's go with averages:

```python
# Average on weekday
weekend = ['Saturday','Sunday']

dotw_smry[ ~dotw_smry['dotw'].isin(weekend)].mean()[0] / dotw_smry[ dotw_smry['dotw'].isin(weekend)].mean()[0]
```

There's roughly 2.1x the traffic on weekdays!
